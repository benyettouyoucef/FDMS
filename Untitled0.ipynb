```python
from sklearn.linear_model import SGDClassifier
import copy
import numpy as np

def loadMovieLens(path='D:/Users/Moi/Desktop/FDMS/FDMS/recodata'):

    # Load data
    prefs={}
    for line in open(path+'/u1.train'):
        (user,movieid,rating,ts)=line.split('\t')
        #print(line)
        prefs.setdefault(user,{})
        prefs[user][movieid]=float(rating)
    return prefs

data  = loadMovieLens()
#print data['1']

```

```python
#Récupérer les utilisateurs + les films + les notes

ttfilm = []
film = []
user = []
notes = []

for u,movie in data.items():
    user.append(u)
    for titre , note in movie.items():
            ttfilm.append(titre)
            notes.append(note)
    film = list(set().union(ttfilm))
    

#minnote = min(notes) 1.0
#maxnote = max(notes) 5.0
#print(len(user)) 943 utilisateurs
print len(film)
```
